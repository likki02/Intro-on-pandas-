```python
# prompt: can you calculate measures of central tendency for salary olumn and measures of dispersion for salry column , also calculate skewness and kutosis ,

import pandas as pd

# Read the CSV file
df = pd.read_csv('Salaries.csv')

# Calculate measures of central tendency
mean_salary = df['salary'].mean()
median_salary = df['salary'].median()
mode_salary = df['salary'].mode()

# Calculate measures of dispersion
variance_salary = df['salary'].var()
std_salary = df['salary'].std()
min_salary = df['salary'].min()
max_salary = df['salary'].max()
range_salary = max_salary - min_salary

# Calculate skewness and kurtosis
skew_salary = df['salary'].skew()
kurt_salary = df['salary'].kurt()

# Print the results
print("Measures of Central Tendency:")
print(f"Mean Salary: {mean_salary:.2f}")
print(f"Median Salary: {median_salary:.2f}")
print(f"Mode Salary: {mode_salary.tolist()}")

print("\nMeasures of Dispersion:")
print(f"Variance of Salary: {variance_salary:.2f}")
print(f"Standard Deviation of Salary: {std_salary:.2f}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Range of Salary: {range_salary}")

print("\nSkewness and Kurtosis:")
print(f"Skewness of Salary: {skew_salary:.2f}")
print(f"Kurtosis of Salary: {kurt_salary:.2f}")
```
